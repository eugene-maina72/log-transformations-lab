## Build a Model with Log-Transformed Features and Target

### Data Preparation

Choose up to 3 of the features you investigated, and set up an X dataframe containing the log-transformed versions of these features as well as a y series containing the log-transformed version of the target.

---

<details>
    <summary style="cursor: pointer"><b>Hint (click to reveal)</b></summary>

If you are planning log transform a feature measured in _years_ (e.g. `YearRemodAdd`) consider shifting the data first. For example, you might subtract 1900 or 1910 from the year, so that a 1% increase in year is closer to meaning 1 year rather than 20 years.

</details>

```python
# Your code here - prepare data for modeling
X['YearRemodAdd'] = X['YearRemodAdd']-1900
X['YearBuilt'] = X['YearBuilt']-(X['YearBuilt'].min() - 2)

X_log_transformed = np.log(X[['YearRemodAdd', 'YearBuilt','2ndFlrSF']].copy())
y_log_transformed = np.log(y)
```

### Modeling

Now build a StatsModels OLS model with a log-transformed target as well as log-transformed features.

```python
# Your code here - build a model
import statsmodels.api as sm
log_transformed_model = sm.OLS(y_log_transformed, sm.add_constant(X_log_transformed)).fit()
```

### Model Evaluation and Interpretation

How did the model perform? How might we interpret its coefficients? Create as many cells as needed.

```python
# Your code here - evaluate the model
log_transformed_model.summary()
```

***
1. **R-squared and Adjusted R-squared**:
   - The **R-squared** value indicates that approximately `59%` of the variance in the log-transformed `SalePrice` is explained by the predictors (`YearBuilt`, `YearRemodAdd`, and `2ndFlrSF`).
   - The **Adjusted R-squared** is slightly lower than the R-squared, which suggests that the model is not overfitting and the predictors are relevant.

2. **P-values for Coefficients**:
   - The p-values for `YearBuilt`, `YearRemodAdd`, and `2ndFlrSF` are all below 0.05, indicating that these predictors are statistically significant in explaining the variance in


3. **Coefficients**:
   - The coefficient for `YearBuilt` indicates that a 1% increase in the log-transformed `YearBuilt` corresponds to a `0.1%` increase in the log-transformed `SalePrice`.
   - Similarly, the coefficients for `YearRemodAdd` and `2ndFlrSF` can be interpreted as the percentage change in `SalePrice` for a 1% change in these predictors.

4. **F-statistic and Prob(F-statistic)**:
   - The **F-statistic** is significant (p-value < 0.05), indicating that the model as a whole is statistically significant and the predictors collectively explain the variance in `SalePrice`.

5. **Model Performance**:
   - Overall, the model performs well, with a high R-squared value and statistically significant predictors.
   - The log transformation of both the predictors and the target appears to have improved the linearity of the relationships, as evidenced by the model's performance metrics.
